```python
from sklearn.metrics import adjusted_rand_score
from sklearn.cluster import KMeans
from sklearn.utils import resample
bootstraps=[]
for i in range(100):
    bootstraps.append(resample(df,random_state=1234))
    
    
adj_rand_i=[]
for i in range(2,9):
    s=[]
    for j in bootstraps:
        model_k=KMeans(n_clusters=i,n_init=10,random_state=1234)
        predict_labels=model_k.fit_predict(j)
        true_labels=model_k.fit_predict(df)
        s.append(adjusted_rand_score(true_labels,predict_labels))
    adj_rand_i.append(s)
```

```python
plt.boxplot(adj_rand_i,labels=[2,3,4,5,6,7,8])
plt.xlabel("Num of Segments")
plt.ylabel("Adjusted Rand Score")
plt.show()
```

```python
seg_stab=[]
for i in range(2,9):
    model_k=KMeans(n_clusters=i,n_init=10,random_state=1234)
    model_k.fit(df)
    labels=model_k.predict(df)
    seg_stab.append(labels)
for i,seg in enumerate(range(2,9)):
    plt.plot(range(2,9), [np.mean(seg_stab[i] == l) for l in seg_stab],marker='o',label='segment{}'.format(seg))
plt.xlabel('Number of Segments')
plt.ylabel('Segment Level Stability')
plt.title('Segment Level Stability Across Solutions (SLSA) Plot')
plt.xticks(range(2,9))
plt.legend()
plt.show()             
```

```python
segment_stability = []

for i in range(2,6):
    segment_similarities=[]
    model_k=KMeans(n_clusters=i,n_init=10,random_state=1234)
    model_k.fit(df)
    segment_similarities= model_k.transform(df).min(axis=1)
    n_similarities=segment_similarities/np.max(segment_similarities)
    segment_stability.append(n_similarities)

plt.boxplot(segment_stability)
plt.xlabel("Segment Number")
plt.ylabel("Segment Stability")
plt.xticks([1,2,3,4],[2,3,4,5])
plt.ylim(0, 1)
plt.title("Segment Level Stability within Solutions")

plt.show()
```

```python
from scipy.stats import entropy
df_1 = []

for k in range(2,9):
    model_k = KMeans(n_clusters=k,n_init=10, random_state=1234)
    model_k.fit(df.values)
    iter_val = model_k.n_iter_
    converged = True
    k_val = k
    k0_val = k
    log_likelihood = -model_k.inertia_
    n_samples, _ = df.shape
    aic = -2 * log_likelihood + 2 * k
    bic = -2 * log_likelihood + np.log(n_samples) * k
    labels = model_k.labels_
    counts = np.bincount(labels)
    probs = counts / float(counts.sum())
    class_entropy = entropy(probs)
    icl = bic - class_entropy
    
    df_1.append((iter_val, converged, k_val, k0_val, log_likelihood, aic, bic, icl))
df_1 = pd.DataFrame(df_1, columns=['iter', 'converged', 'k', 'k0', 'logLik', 'AIC', 'BIC', 'ICL'])

print(df_1)
```

```python
plt.plot(df_1["k"], df_1["AIC"], marker='o', label='AIC')
plt.plot(df_1["k"], df_1["BIC"], marker='o', label='BIC')
plt.plot(df_1["k"], df_1["ICL"], marker='o', label='ICL')

plt.xlabel('Number of Segments')
plt.ylabel('Value of Information Criteria')
plt.title('Information Criteria (AIC, BIC, ICL)')
plt.legend()

plt.show()
```
